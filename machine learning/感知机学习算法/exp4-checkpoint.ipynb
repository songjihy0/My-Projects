```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataSet = []
x = []
y = []
        
def Perceptron_orgin():
    # 训练集

    with open(".\perceptron_data.txt", 'r') as f:
        for line in f.readlines():
            list = []
            for i in line.split():
                list.append(float(i))
            dataSet.append(list) 
#     print("data:",dataSet)
    
    for line in dataSet:

        if line[2] == 1.0:
            y.append(1.0)
        else: y.append(-1.0)
        line.pop()
        x.append(line)
    
    # 初始化w，b
    w = np.zeros(2).reshape(1,2)
    θ = 0

    while True:
        num_errors = 0
        for i in range(98):
            # 判断是否误分类
            if y[i] * (np.dot(w, x[i]) + θ) <= 0:
                # 利用随机梯度下降算法，更新w，θ
                w[0][0] += y[i] * x[i][0]
                w[0][1] += y[i] * x[i][1]
                θ += y[i]
                # 误分类个数加1
                num_errors += 1
                print("迭代更新:",w, θ)
        # M中为空训练结束
        if num_errors == 0:
#             print(w, b)
            break
    
    # 画出超平面（二维中即一条直线）
    x0 = np.arange(-4,5)
    y0 = -(w[0][0] / w[0][1]) * x0 - θ / w[0][1]
    plt.plot(x0, y0, 'r')
    # 画出训练集中的点
    
    for i in range(98):
        if y[i]==1.0:
            plt.scatter(x[i][0], x[i][1], c='b')
        else:plt.scatter(x[i][0], x[i][1], c='r')

    plt.show()

if __name__ == '__main__':
    Perceptron_orgin()
```
